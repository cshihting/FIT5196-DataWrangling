##### * The context-independent and context-dependent (with the threshold set to %98) stop words <br> must be removed from the vocab. The stop words list (i.e,  stopwords_en.txt ) provided in the<br> zip file must be used. 
移除stopword

```python
# context-dependent (with the threshold set to %98)這個還沒處理

# create an empty list
stopwords_list = []

# open and read stopwords_en.txt
with open('stopwords_en.txt', 'r') as f:
    # read line by line
    for word in f.readlines():
        # remove newline
        word = word.strip()
        # append each stop word into stopwords_list
        stopwords_list.append(word)
# stopwords_list

# to create a tokens list without stopwords
filtered_tokens = [token for token in tokens if token not in stopwords_list]
filtered_tokens # tokens list without stopwords
```

```python
# list of alphabets
alphabets = ['a', 'b', 'd', 'e', 'i', 'j', 'o']

# function that filters vowels
def filterVowels(alphabet):
    vowels = ['a', 'e', 'i', 'o', 'u']

    if(alphabet in vowels):
        return True
    else:
        return False

filteredVowels = filter(filterVowels, alphabets)

result = ''
print('The filtered vowels are:')
for vowel in filteredVowels:
    result += vowel+' '
print(result)
```

```python
# 卡題
# regex for the Camelcase
#r'(?<!\.\s)\b[A-Z][a-z]*\b' # get the capital tokens appeared in the middle
#camel_case = r"([A-Z][a-z]+[A-Z][a-z]+)"

# regex for getting those capital tokens in the middle of the sentences 
capita_tokens = r'(?<!\.\s)(\b[A-Z](?:[A-Z]*|[a-z]*)\b|\b[A-Z][a-z]+[A-Z][a-z]+\b)'
# capita_first_tokens = r'(?:^|(?:\W\s))([A-Z]\w+)'
# 包含句首那個字也算入 r'(?:(?<!\.\s)(\b[A-Z](?:[A-Z]*|[a-z]*)\b|\b[A-Z][a-z]+[A-Z][a-z]+\b))|(?:(?:^|(?:\W\s))([A-Z]\w+))'

match_capital = re.findall(capital_tokens, clean_string)
# match_capital

# create an empty string
result = ''
# check each token in tokens
for token in tokens:
    # if that token not in the middle of the sentences
    if token not in match_capital:
        # make it lowercase and then add into result
        result += token.lower() + ' '
    else:
        # or directly add into result
        result += token + ' '
print(result)

# # 順序會亂掉，但留下來的是對的值
# match_capital_set = set(match_capital)
# tokens_set = set(tokens)
# print(tokens_set-match_capital_set)
```

##### * First 200 meaningful bigrams (i.e., collocations) must be included in the vocab.
剩下的字中，前200的bigram<br>
ex: I am/ soy milk/ google map 兩個兩個字

```python
# Extracting from a text a list of n-gram can be easily accomplished with function ngram()
from nltk.util import ngrams
# NLTK provides a built-in function FreqDist to compute this distribution directly from a set of word tokens.
from nltk.probability import *

# create an empty string
non_sign = ''
# check each token in tokens
for each in tokens:
    # if each token is an alphabet
    if each.isalpha():
        # add it into non_sign
        non_sign += each + " "
        
# get bigrams
bigrams = ngrams(non_sign.split(), n = 2)
# the frequency of eahc bigram
fdbigram = FreqDist(bigrams)
```

```python
# get 200 higher frequency (much more meaningful) bigrams and sort
sorted(fdbigram.most_common(n=200))
```

```python
bigram_measures = nltk.collocations.BigramAssocMeasures()
finder = nltk.collocations.BigramCollocationFinder.from_words(non_sign.split())
finder.nbest(bigram_measures.pmi, 200)
```

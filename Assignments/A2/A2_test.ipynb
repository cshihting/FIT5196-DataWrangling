```python
import numpy as np

# get some specific columns I need
dist_df = pd.DataFrame(df1[['origin_latitude', 'origin_longitude', 'destination_latitude', 'destination_longitude', 'journey_distance']], columns=['origin_latitude', 'origin_longitude', 'destination_latitude', 'destination_longitude', 'journey_distance'])

# the function for calculating distance
def calculate_dist(row):
    dist = 6378 * acos(sin(radians(row['origin_latitude']))*sin(radians(row['destination_latitude'])) + cos(radians(row['origin_latitude']))*cos(radians(row['destination_latitude']))*cos(radians(row['origin_longitude']) - radians(row['destination_longitude'])))
    distance = round(dist, 3)
    return distance

# axis=1 or columns: apply function to each row.
dist_series = dist_df.apply(calculate_dist, axis=1)
# add the result of calculation into data frame named 'cal_result'
dist_df['cal_result'] = pd.Series(dist_series, index=dist_df.index)
# find which journey_distance is different from cal_result
dist_df['comparison'] = np.where((dist_df['journey_distance'] == dist_df['cal_result']), 'same', 'diff')
dist_df
```

### Travel time (deal with travel_time)

```python
from datetime import datetime
from datetime import timedelta
import numpy as np

# the function for calculating the time interval and convert it to minute
def cal_time_interval(row):
    FMT = '%H:%M:%S' # format
    timedelta = datetime.strptime(row['delivery_time'], FMT) - datetime.strptime(row['departure_time'], FMT)
    mins = timedelta.seconds/60
    result = round(mins, 2)
    return result

# call the function (the type of the outcome is series)
# axis=1 or ‘columns’: apply function to each row.
time_series = df.apply(cal_time_interval, axis=1)

# add the result of calculation into data frame and name it 'cal_time'
df['cal_mins'] = pd.Series(time_series, index=df.index)

# display the data frame
df.head(n=3)
```

```python
df.loc[[6171]]
```

```python
# find which travel_time is too far from the actual result
# if (travel_time - cal_mins) < 0.02, then show close; or show too_much
df['comparison'] = np.where((df['travel_time'] - df['cal_mins']) <= 0.02, 'close', 'too_much')
df.head(n=3)
```

```python
# # take a look
# df[df['comparison']=='too_much']
```

```python
# replace the travel_time where its comparison = 'too_much' with the actual mins (cal_mins)
df['travel_time'] = np.where(df['comparison'] == 'too_much', df['cal_mins'], df['travel_time'])
df[df['comparison']=='too_much'].head(n=2)
```

### Speed (deal with drone_type)

```python
# the function for calculating the speed, 
# exculding the time of takeOff and Landing (total: 10 mins)
def cal_speed(row):
    speed = row['journey_distance']/row['travel_time']
    return speed

# call the function (the type of the outcome is series)
# axis=1 or ‘columns’: apply function to each row.
minus_series = df.apply(cal_speed, axis=1)

# add the result into data frame and name it 'on_route'
df['speed'] = pd.Series(minus_series, index=df.index)

# display the data frame
df
```

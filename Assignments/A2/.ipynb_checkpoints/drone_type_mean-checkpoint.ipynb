```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("29286875.csv")

# change the column names
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.head(n=1)
```

```python
# the function for calculating speed (distance/time)
def cal_speed(row):
    # exclude takeOff time (5 mins) to make the speed more accurate >> as a result, not more accurate
    #speed = row['journey_distance'] / (row['travel_time'] - 5)
    speed = row['journey_distance'] / row['travel_time']
    return speed


# axis=1 or ‘columns’: apply function to each row.
speed_series = df.apply(cal_speed, axis=1)

# add the result of calculation into data frame named 'on_route'
df['speed'] = pd.Series(speed_series, index=df.index)

# # find which journey_distance is different from cal_result
# post_fare_df['comparison'] = np.where((dist_df['journey_distance'] == dist_df['cal_result']), 'same', 'diff')
df
```

#### try to get the correct values for those NaN


```python
# extract rows whose values are NaN 
drone_null_df = df[df['drone_type'].isnull()]

# convert series type to list type (get the speed for comparison later)
drone_null_list = drone_null_df['speed'].tolist()

# convert the series of three types of drone to a list
speed_list = df.groupby('drone_type')['speed'].mean().tolist()
```

```python
# get two lists (one is minmum gap list; another is all gap result)
# 一個個相減去找最小差，推測該速度應該屬於哪個型號
# 在此先分成兩個list，一個是最小差的list，一個是所有相減的結果
gap_min_list = [] # to store the value of the minmum gap 
all_gap_list = [] # to store the values of all gap
for each in drone_null_list: # each speed of drone type whose value is NaN
    temp_list = [] # create an empty list
    for speed in speed_list: # the speed of three types
        gap = np.abs(each-speed) # get the positive gap 
        all_gap_list.append(gap) # add into all_gap_list
        temp_list.append(gap) # add 3 gap of 3 type into temp_list
    gap_min_list.append(min(temp_list)) # add the minmum gap into gap_min_list

# check the results
print("The minmum gap list:\n", gap_min_list)
print("\nThe all gap list:\n", all_gap_list)
```

```python
# reshape the all_gap_list to compare to each value in gap_min_list
all_array = np.reshape(all_gap_list, (10, 3))
```

```python
# get the drone type list 
type_list = [] # create a type list for those whose drone type is NaN
for each in gap_min_list: # each in gap_min_list
    for row in range(len(all_array)): # row by row in all_array (total 10)
        for value in range(3): # each value in a row (3 values per row)
            if each == all_array[row][value]: # get the matched value
                type_list.append(value+1) # add into type_list and give the correct type number

# check the list
print(type_list)
```

#### try to replace NaN with the correct values

```python
# copy the dataframe
df_copy = df.copy()
```

```python
# fill NaN with 0 to that new copy dataframe
df_copy[['drone_type']] = df_copy[['drone_type']].fillna(value=0)
```

```python
# check (NaN has been replaced by 0 already)
df_copy[df_copy['drone_type'].isnull()]
```

```python
# convert the column to a list
drone_type_list = df_copy['drone_type'].tolist()
```

```python
# count how many values are 0 in drone_type_list (now is 10)
drone_type_list.count(0)
```

```python
# reverse the list (can pop values from the list properly later on)
type_list.reverse()
```

```python
# replace the 0 with the correct drone type number
for each in range(len(drone_type_list)):
    if drone_type_list[each] == 0: # if the value in drone_type_list is 0
        drone_type_list[each] = type_list.pop() # pop the correct one from type_list to replace 0 in drone_type_list
drone_type_list.count(0) # count how many values are 0 in drone_type_list (now is 0)
```

```python
# add the list into data frame (df) and name it 'correct_drone_type'
df['correct_drone_type'] = pd.Series(drone_type_list, index=df.index)
```

```python
# replace NaN with the correct type number
df['drone_type'] = np.where(df['drone_type'].isnull(), df['correct_drone_type'], df['drone_type'])
```

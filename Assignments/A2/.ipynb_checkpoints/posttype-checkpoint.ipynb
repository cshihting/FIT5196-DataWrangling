```python
import pandas as pd

df = pd.read_csv("29286875.csv")

# change the column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.head(n=10)
```

```python
df['id'] = df['id'].str.replace('ID', '')
df.head(n=10)
```

```python
# find id is only start with 1 or 5 (the remaining are NaN)
df.loc[(df['id'].str[0] != "1") & (df['id'].str[0] != "5")]
```

```python
# ID start with 5, post_type is 1
df.loc[(df['id'].str[0]) == "5"]
```

```python
type(df.loc[0]['post_type'])
```

```python
df.loc[(df['id'].str[0] == "5") & (df['post_type'] == 0.0)]
```

```python
# ID start with 1, post_type is 0
df.loc[(df['id'].str[0]) == "1"]
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline # 'magic line' to make sure that graphs are shown inline in the notebook
```

```python
# id v.s. post_type
plt.scatter(df['id'], df['post_type'])
plt.title("The Price of The Travel Distance")
plt.xlabel("Fare")
plt.ylabel("Travel Distance")
plt.show()
```

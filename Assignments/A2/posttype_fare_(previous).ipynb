```python
import pandas as pd

df = pd.read_csv("29286875.csv") #, parse_dates=[0])
# print(df1)
```

```python
# change the column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
df.head(n=3)
```

```python
df.isnull().sum()
```

```python
posttype_df = pd.DataFrame(df[['post_type', 'departure_time', 'delivery_fare']], columns=['post_type', 'departure_time', 'delivery_fare'])
posttype_df #.between_time('06:00:00', '11:59:59')
```

```python
# make 'departure_time' as index and convert index type to DatetimeIndex
posttype_df.set_index('departure_time', inplace=True)
posttype_df.index = pd.to_datetime(posttype_df.index)
posttype_df
```

```python
# posttype_df.reset_index()
```

```python
morning_df = posttype_df.between_time('06:00:00', '11:59:59')
afternoon_df = posttype_df.between_time('12:00:00', '20:59:59')
night_df = posttype_df.between_time('21:00:00', '05:59:59')
```

```python
# extract rows whose values are NaN 
posttype_null_df = posttype_df[posttype_df['post_type'].isnull()]
# posttype_null_df

# convert series type to list type (get the fare for comparison later)
posttype_null_fare_list = posttype_null_df['delivery_fare'].tolist()
# posttype_null_list

# convert the series of 2 post types (morning) to a list
morning_list = morning_df.groupby('post_type')['delivery_fare'].mean().tolist()
afternoon_list = afternoon_df.groupby('post_type')['delivery_fare'].mean().tolist()
night_list = night_df.groupby('post_type')['delivery_fare'].mean().tolist()
```

```python
import numpy as np
# merge 3 lists into one
all_time_fare_list = morning_list + afternoon_list + night_list
```

```python
# get two lists (one is minmum gap list; another is all gap result)
# 一個個相減去找最小差，推測該速度應該屬於哪個型號
# 在此先分成兩個list，一個是最小差的list，一個是所有相減的結果
post_gap_min_list = [] # to store the value of the minmum gap 
post_all_gap_list = [] # to store the values of all gap
for each in posttype_null_fare_list: # each fare of post type whose value is NaN
    temp_list = [] # create an empty list
    for posttype in all_time_fare_list: # the fare of 2 post types
        gap = np.abs(each-posttype) # get the positive gap 
        post_all_gap_list.append(gap) # add into all_gap_list
        temp_list.append(gap) # add 6 gap (2 types for each sections) into temp_list
    post_gap_min_list.append(min(temp_list)) # add the minmum gap into gap_min_list

# check the results
print("The minmum gap list:\n", post_gap_min_list)
print("\nThe all gap list:\n", post_all_gap_list)
```

```python
# reshape the all_time_list to compare to each value in posttype_null_list
all_array = np.reshape(post_all_gap_list, (20, 6))
all_array
```

```python
# get the drone type list 
two_type_list = [] # create a type list for those whose drone type is NaN
for each in post_gap_min_list: # each in gap_min_list
    for row in range(len(all_array)): # row by row in all_array (total 20)
        for value in range(6): # each value in a row (6 values per row)
            if each == all_array[row][value]: # get the matched value
                if value % 2 == 0: # normal post type
                    two_type_list.append(0) # add into type_list and give the correct post type number
                else:
                    two_type_list.append(1)

# check the list
print(two_type_list)
```

```python
# copy the dataframe
df_copy = df.copy()
```

```python
# fill NaN with 2 to that new copy dataframe
df_copy[['post_type']] = df_copy[['post_type']].fillna(value=2)
```

```python
# check (NaN has been replaced by 2 already)
df_copy[df_copy['post_type'].isnull()]
```

```python
# convert the column to a list
post_type_list = df_copy['post_type'].tolist()
```

```python
# count how many values are 2 in post_type_list (now is 20)
post_type_list.count(2)
```

```python
# reverse the list (can pop values from the list properly later on)
two_type_list.reverse()
```

```python
# replace the 2 with the correct post type number
for each in range(len(post_type_list)):
    if post_type_list[each] == 2: # if the value in post_type_list is 2
        post_type_list[each] = two_type_list.pop() # pop the correct one from two_type_list to replace 2 in post_type_list
post_type_list.count(2) # count how many values are 2 in drone_type_list (now is 0)
```

```python
# add the list into data frame (df) and name it 'correct_post_type'
df['correct_post_type'] = pd.Series(post_type_list, index=df.index)
```

```python
# replace NaN in post_type with the correct post type number
df['post_type'] = np.where(df['post_type'].isnull(), df['correct_post_type'], df['post_type'] )
```

```python
# check (Now there is no NaN value in post_type)
df.isnull().sum()
```
